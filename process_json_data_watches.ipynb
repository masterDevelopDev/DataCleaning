### Watches documents

```python
import os

# Specify the directory you want to search
directory = r'C:\Users\delll\Downloads\metadata\ListData_Watches'

# Initialize a list to hold all the file paths
json_files = []

# Walk through directory
for dirpath, dirnames, files in os.walk(directory):
    for file in files:
        if file.endswith(".json"):
            json_files.append(os.path.join(dirpath, file))

json_files
```

```python
list_df_watches = []
for json_file in json_files:
    df = pd.read_json(json_file)
    df['FileLeafRef'] = df['FileLeafRef'].fillna('')
    df = df[df['FileLeafRef'].str.contains('.pdf')].drop_duplicates()
    df['FileLeafRef'] = df['FileLeafRef'].apply(lambda x: x.replace('.aspx', ''))
    list_df_watches.append(df)
```

```python
df_concatenated = pd.concat(list_df_watches, ignore_index=True).drop_duplicates()

def change_year_format(year_str):
    if year_str:
        return int(year_str.replace(',', ''))
    else:
        return ''
    
def change_str_to_int(string):
    if string:
        return int(string)
    else:
        return ''

df_concatenated['AnneePublication'] = df_concatenated['AnneePublication'].apply(change_year_format)
df_concatenated['Modified'] = df_concatenated['Modified'].astype(str)
df_concatenated['Created'] = df_concatenated['Created'].astype(str)
df_concatenated['CadrComptNbre'] = df_concatenated['CadrComptNbre'].apply(change_str_to_int)
df_concatenated['CadrGuichetNbre'] = df_concatenated['CadrGuichetNbre'].apply(change_str_to_int)
df_concatenated['CadrNbre'] = df_concatenated['CadrNbre'].apply(change_str_to_int)
df_concatenated['CornNbre'] = df_concatenated['CornNbre'].apply(change_str_to_int)
df_concatenated['CourNbre'] = df_concatenated['CourNbre'].apply(change_str_to_int)
df_concatenated['LuneVisNbre'] = df_concatenated['LuneVisNbre'].apply(change_str_to_int)
df_concatenated
```

```python
remaining_columns = ['FileLeafRef',  'MontreType', 'AiguForm', 'AiguOrn', 'BoitOrn', 'BoitForm',
       'BoitFlanc', 'BoucForm', 'BoucOrn', 'BracStructure', 'BoucType',
       'BracMat', 'BracOrn', 'CadrCheminFer', 'CadrChiffre', 'CadrComptNbre',
       'CadrEnchev', 'CadrForme', 'CadrGuichetNbre', 'CadrNbre', 'CadrOrn',
       'CadrRehaut', 'Corne', 'CornNbre', 'CourCabo', 'CourDispo', 'CourNbre',
       'CourProt', 'Couleur', 'IndeForme', 'IndeOrn', 'LuneChiffre',
       'LuneForme', 'LuneOrn', 'LunePlongee', 'LuneVisNbre', 'MecaFonction',
       'MecaMvmtVisible', 'Ornement', 'PousForme', 'PousNbre', 'VerrProt',
       'VisForme', 'AnneePublication', 
       'Modified', 'Depot', 'Created',
       'Author', 'Editor', 'Region']

depot_region_mappings = {
    "CN": 'Chine',
    "HK": 'Hong Kong',
    "IFPI": 'Suisse',
    "INPI": 'France',
    "OHMI": 'UE',
    "OMPI": 'Monde',
    "USPTO": 'USA',
    "": ''
}
df_concatenated['Region'] = df_concatenated['Depot'].map(depot_region_mappings)
df_concatenated = df_concatenated[remaining_columns]
df_concatenated
```

```python
file_path = 'Downloads/original_metadata.xlsx'

# Read the Excel file
df_watches_metadata = pd.read_excel(file_path, sheet_name='Watches')
df_watches_metadata = df_watches_metadata[['Level 0', 'Level 1', 'Level 2', 'Level 3', 'Alias Catalog']].fillna('')
df_watches_metadata
```

```python
def concatenate_strings(row, columns):
    list_strings = [row[col] for col in columns if row[col]][:-1]
    return '.'.join(list_strings).lower()
```

```python
level_columns = ['Level 0', 'Level 1', 'Level 2', 'Level 3']
df_watches_metadata['paths'] = df_watches_metadata.apply(lambda row: concatenate_strings(row, level_columns), axis=1)
df_watches_metadata
```

```python
metadata_mapping = df_watches_metadata[['Alias Catalog', 'paths']].drop_duplicates()
metadata_mapping
```

```python
metadata_mapping.loc[metadata_mapping['Alias Catalog'] == 'BracRangNbre', 'paths'] = 'watches.bracelet.rangnombre'
metadata_mapping.loc[metadata_mapping['Alias Catalog'] == 'CadrComptNbre', 'paths'] = 'watches.cadran.comptnbre'
metadata_mapping.loc[metadata_mapping['Alias Catalog'] == 'CadrGuichetNbre', 'paths'] = 'watches.cadran.guichetnbre'
metadata_mapping.loc[metadata_mapping['Alias Catalog'] == 'CadrNbre', 'paths'] = 'watches.cadran.nbre'
metadata_mapping.loc[metadata_mapping['Alias Catalog'] == 'CornNbre', 'paths'] = 'watches.corne.nbre'
metadata_mapping.loc[metadata_mapping['Alias Catalog'] == 'CourNbre', 'paths'] = 'watches.couronne.nbre'
metadata_mapping.loc[metadata_mapping['Alias Catalog'] == 'LuneVisNbre', 'paths'] = 'watches.lune.visnbre'
metadata_mapping.loc[metadata_mapping['Alias Catalog'] == 'PousNbre', 'paths'] = 'watches.poussoir.nbre'
```

```python
d = {'Alias Catalog': ['AnneePublication', 'Modified', 'Created', 'Depot', 'Author', 'Editor', 'FileLeafRef', 'Region'], 
     'paths': ['year', 'modified', 'created', 'depot', 'author', 'editor', 'filename', 'region']}

df_structu_metadata = pd.DataFrame(data=d)
df_structu_metadata
```

```python
metadata_names_df = pd.concat([df_structu_metadata, metadata_mapping], ignore_index=True)
metadata_names_df = metadata_names_df.replace('LunePlongée', 'LunePlongee')
metadata_names_df
```

```python
metadata_names_dict = dict(zip(metadata_names_df['Alias Catalog'], metadata_names_df['paths']))
del metadata_names_dict['Ornement']  
metadata_names_dict
```

```python
orn_val_maps_df = df_watches_metadata[df_watches_metadata['Level 1'] == 'Ornement'][['Level 3', 'paths']].groupby('Level 3')['paths'].agg(list).reset_index()
ornement_value_mappings = dict(zip(orn_val_maps_df['Level 3'], orn_val_maps_df['paths']))
ornement_value_mappings
```

```python
def merge_dictionaries(list_of_dicts):
    merged_dicts = {}
    for d in list_of_dicts:
        if d['filterId'] in merged_dicts:
            # Extend the list of values if the path is already present
            merged_dicts[d['filterId']].extend(d['values'])
        else:
            # Otherwise, add the new path and its values
            merged_dicts[d['filterId']] = d['values']
    # Remove duplicates from the values lists
    for path, values in merged_dicts.items():
        merged_dicts[path] = list(set(values))
    # Convert the merged dictionary back to a list of dictionaries
    merged_list_of_dicts = [{'filterId': path, 'values': values} for path, values in merged_dicts.items()]
    return merged_list_of_dicts
```

```python
def get_filter_values(row, metadata_names_dict, ornement_value_mappings, cols_filter_values, cols_struct_metadata):
    dict_file = {}
    list_filter_values_paths = []
    for column in cols_struct_metadata:
        dict_file[metadata_names_dict[column]] = row[column]
    for column in cols_filter_values:
        values_taken = row[column]
        if values_taken != '':
            if column != 'Ornement':
                filter_path = metadata_names_dict[column]
                if type(values_taken) == int:
                    list_filter_values_paths.append({'filterId': filter_path, 
                                                     'values': [values_taken]})
                else:
                    list_values = values_taken.split('|')
                    list_filter_values_paths.append({'filterId': filter_path, 
                                                     'values': list_values})
            else:
                dict_path_orn = {}
                list_values = values_taken.split('|')
                for value_orn in list_values:
                    try:
                        path_orn = ornement_value_mappings[value_orn][0]
                        if path_orn not in dict_path_orn.keys():
                            dict_path_orn[path_orn] = [value_orn]
                        else:
                            dict_path_orn[path_orn] = dict_path_orn[path_orn] + [value_orn]
                            for filter_path in dict_path_orn.keys():
                                list_filter_values_paths.append({'filterId': filter_path, 
                                                                 'values': dict_path_orn[filter_path]})
                    except KeyError:
                        print(f"The value {value_orn} found in the file {row['FileLeafRef']} does not exist in the metadata.")
    list_filter_values_paths = merge_dictionaries(list_filter_values_paths)
    dict_file['filterValues'] = list_filter_values_paths            
    return dict_file
        

columns_filter_values = ['MontreType', 'AiguForm', 'AiguOrn', 'BoitOrn', 'BoitForm', 'BoitFlanc', 
                  'BoucForm', 'BoucOrn', 'BracStructure', 'BoucType', 'BracMat', 'BracOrn', 'CadrCheminFer', 
                  'CadrChiffre', 'CadrComptNbre', 'CadrEnchev', 'CadrForme', 'CadrGuichetNbre', 'CadrNbre', 'CadrOrn', 
                  'CadrRehaut', 'Corne', 'CornNbre', 'CourCabo', 'CourDispo', 'CourNbre', 'CourProt', 'Couleur', 
                  'IndeForme', 'IndeOrn', 'LuneChiffre', 'LuneForme', 'LuneOrn', 'LunePlongee', 'LuneVisNbre', 
                  'MecaFonction', 'MecaMvmtVisible', 'Ornement', 'PousForme', 'PousNbre', 'VerrProt', 'VisForme']

columns_struct_metadata = ['FileLeafRef', 'AnneePublication', 'Modified', 'Depot', 'Created', 'Author', 'Editor', 'Region']

df_concatenated['dict_filter_values'] = df_concatenated.apply(lambda row: get_filter_values(row, 
                                                                                            metadata_names_dict, 
                                                                                            ornement_value_mappings,
                                                                                            columns_filter_values,
                                                                                            columns_struct_metadata), axis=1)
```

```python
import json

# Specify the filename
filename = 'watches_documents_metadata.json'

# Write JSON data to a file
with open(filename, 'w', encoding='utf-8') as file:
    json.dump(list(df_concatenated['dict_filter_values']), file, ensure_ascii=False, indent=4)
```
